```python
from ete3 import PhyloTree, Nexml, Tree, faces, AttrFace, TreeStyle

SpTree = Tree("../Data/OFResults/OFSpeciesTree.nw", format=0)

#midpoint root the tree
SpTree.set_outgroup(SpTree.get_midpoint_outgroup())

SpTree.ladderize()

#Name all the internal nodes
edge = 0
for node in SpTree.traverse():
    if not node.is_leaf():
        node.name = "n%d" %edge
        edge += 1

print(SpTree.get_ascii(attributes=["name", "species"], show_internal=False))
```

```python
#Change layout to display internal node names
def my_layout(node):
    if node.is_leaf():
         # If terminal node, draws its name
         name_face = AttrFace("name")
    else:
         # If internal node, draws label with smaller font size
         name_face = AttrFace("name", fsize=1)
    # Adds the name face to the image at the preferred position
    faces.add_face_to_node(name_face, node, column=0, position="branch-bottom")

ts = TreeStyle()
# Do not add leaf names automatically
ts.show_leaf_name = False
# Use my custom layout
ts.layout_fn = my_layout

#SpTree.render("mytree.png", w=183, units="mm")

#t = SpTree
#t.populate(10, random_dist=True)
#ts = TreeStyle()
#ts.show_leaf_name = True
#ts.show_branch_length = True
#ts.show_branch_support = True
SpTree.show(tree_style=ts)
```
