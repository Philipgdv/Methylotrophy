Create a mapping data frame linking each OrthoFinder OG to the eOG used by eggNOG

```python
emapper = pd.read_csv("../Data/input/MM_jbrjl6fg.emapper.annotations.tsv", sep ="\t", comment = "#", header = None)
emapper.columns = ["query","seed_ortholog","evalue","score","eggNOG_OGs","max_annot_lvl","COG_category","Description","Preferred_name","GOs","EC","KEGG_ko","KEGG_Pathway","KEGG_Module","KEGG_Reaction","KEGG_rclass","BRITE","KEGG_TC","CAZy","BiGG_Reaction","PFAMs"]

OFOG_Annot = pd.DataFrame()
OFOG_Annot["OFOG"] = emapper["query"].str.split(".",expand=True)[0]
OFOG_Annot["eOG"] = emapper["eggNOG_OGs"].str.split("@",expand=True)[0]

OFOG_Annot
```

```python
emapper = pd.read_csv("../Data/input/MM_jbrjl6fg.emapper.annotations.tsv", sep ="\t", comment = "#", header = None)
emapper.columns = ["query","seed_ortholog","evalue","score","eggNOG_OGs","max_annot_lvl","COG_category","Description","Preferred_name","GOs","EC","KEGG_ko","KEGG_Pathway","KEGG_Module","KEGG_Reaction","KEGG_rclass","BRITE","KEGG_TC","CAZy","BiGG_Reaction","PFAMs"]

OFOGs = pd.read_csv("../Data/input/Proteomes/OrthoFinder/Results_Dec03/Orthogroups/Orthogroups.GeneCount.tsv", sep ="\t", comment = "#", index_col = 0, header = 0)

OFOG_Annot = pd.DataFrame()
OFOG_Annot["OFOG"] = emapper["query"].str.split(".",expand=True)[0]
OFOG_Annot["eOG"] = emapper["eggNOG_OGs"].str.split("@",expand=True)[0]

OFOGs.astype(int)

OFOGs = OFOGs.merge(OFOG_Annot, how = "left", left_index=True, right_on = "OFOG").set_index("OFOG").drop("Total", axis = 1)

eOG = OFOGs.eOG.dropna().unique()

for OFOG in OFOGs[OFOGs.eOG.isna()].index:
    OFOGs.eOG.loc[OFOG] = OFOG

OFOGs
```

```python
OFOGs[OFOGs.eOG.duplicated()]
```

```python
Root_OGMembers = pd.read_csv("../Data/input/EggNOG/1_members.tsv", sep ="\t", comment = "#", header = None, index_col = 1)
Root_OGMembers = Root_OGMembers[Root_OGMembers.index.isin(eOG)]
Root_OGMembers.loc[:,5] = Root_OGMembers.loc[:,5].str.split(",", expand = False)

Root_OGMembers_temp = pd.DataFrame(Root_OGMembers.loc[:,5].explode()).reset_index()
Root_OGMembers_temp.columns = ["eOG","Species"]
Root_OGMembers_temp["Presence"] = 1.0
Root_OGMembers_temp.groupby("Species")
Root_OGMembers = Root_OGMembers_temp.pivot(index='eOG', columns='Species', values='Presence').fillna(0)#.iloc[:,"1000565"]

CombinedDF = OFOGs.merge(Root_OGMembers, how = "outer", left_index=True, right_index=True).fillna(0)
CombinedDF
```

Create Species index DF to subset above.

```python
SpeciesIndexDF = pd.read_csv("../Data/input/SpeciesIndexDF.tsv", sep ="\t", comment = "#", header = 0, index_col = "Proteome Id", dtype=str)["Organism"]
SpeciesIndexDF = SpeciesIndexDF.str.replace("Ca. ","").str.split(" ", n=1, expand = True)[0]

#manual correction
SpeciesIndexDF["UP000321408"] = "Lokiarchaeota"
SpeciesIndexDF["UP000625964"] = "Lokiarchaeota"
SpeciesIndexDF["UP000721749"] = "Heimdallarchaeota"
SpeciesIndexDF["UP000625964"] = "Lokiarchaeota"
SpeciesIndexDF["UP000721749"] = "Heimdallarchaeota"
SpeciesIndexDF["UP000775429"] = "Heimdallarchaeota"

SpeciesIndexDF = pd.DataFrame(SpeciesIndexDF)
SpeciesIndexDF["Kingdom"] = "Archaea"
SpeciesIndexDF["Lineage"] = "Asgard"
SpeciesIndexDF.columns = ["Species","Kingdom","Lineage"]

TaxidInfo = pd.read_csv("../Data/input/EggNOG/e5.taxid_info.tsv", sep ="\t", header = 0, index_col = "# Taxid", dtype=str)
TaxidInfo.Lineage = TaxidInfo.Lineage.str.split(",", n=1, expand = True)[0]
TaxidInfo = TaxidInfo.drop(["Rank","Taxid Lineage"], axis = 1)
TaxidInfo.columns = ["Species","Kingdom","Lineage"]

SpeciesIndexDF_grouped = SpeciesIndexDF.append(TaxidInfo)
SpeciesIndexDF_grouped.index = SpeciesIndexDF_grouped.index.astype(str)
SpeciesIndexDF_grouped
```

```python
Lineages = np.concatenate([SpeciesIndexDF_grouped[SpeciesIndexDF_grouped.Kingdom == "Archaea"].Lineage.unique(),
SpeciesIndexDF_grouped[SpeciesIndexDF_grouped.Kingdom == "Bacteria"].Lineage.unique()])
```

```python
ByLineage = pd.DataFrame(index = CombinedDF.index)
for lineage in Lineages:
    ByLineage[lineage] = CombinedDF.loc[:,SpeciesIndexDF_grouped.index[SpeciesIndexDF_grouped.Lineage == lineage]].mean(axis = 1)
    
ByLineage.transpose()
```

mask = SpeciesIndexDF_grouped[SpeciesIndexDF_grouped.Kingdom == "Archaea"].index.values
CombinedDF.loc[:,CombinedDF.columns.isin(mask)]

```python
Data = ByLineage.transpose()
Data
```

```python
labels = Data.columns
width = 1

fig, ax = plt.subplots()

for lineage in Data.index:
    ax.bar(labels, Data.loc[lineage,:]/sum(Data.loc[lineage,:]), width=1)

ax.set_ylabel('Prevelance')
ax.set_title('OG')
ax.legend()

plt.xticks([])
plt.yticks([])
plt.show()
```
