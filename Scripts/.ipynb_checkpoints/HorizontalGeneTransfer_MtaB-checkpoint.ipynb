```python
COG1410_combined = PhyloTree("../HorizontalGeneTransferAnalysis/COG1410_combined.nwk", format=1, sp_naming_function=lambda name: ncbi.get_lineage(name.split('.')[0])[3])

for lf in COG1410_combined.iter_leaves():
    lf.name = ncbi.get_lineage(lf.name.split('.')[0])[3]

def collapsed_leaf(node):
    if len(node2labels[node]) == 1:
        return True
    else:
        return False

def name_internal_nodes(T):
    for node in T.traverse():
        if node.is_leaf()==False:
            # list names of leaves
            leaf_names=[leaf.name for leaf in node.iter_leaves()]
            names_unique = list(set(leaf_names))
            # if all leaves have the same name, give that name to the node
            if (len(names_unique)==1):
                node.name = names_unique[0]

name_internal_nodes(COG1410_combined)

# Cache node content
node2labels = COG1410_combined.get_cached_content(store_attr="name")
# Collapse tree
t2 = Tree(COG1410_combined.write(is_leaf_fn=collapsed_leaf, quoted_node_names=True), quoted_node_names=True)

for lf in t2.iter_leaves():
    lf.name = ncbi.get_taxid_translator([lf.name])
    if lf.name == ncbi.get_taxid_translator([ncbi.get_lineage(1936272)[3]]):
        lf.set_style(ns1)
    elif lf.name == ncbi.get_taxid_translator([ncbi.get_lineage(1706441)[3]]):
        lf.set_style(ns1)

t2.ladderize()
t2.render("%%inline")
```
