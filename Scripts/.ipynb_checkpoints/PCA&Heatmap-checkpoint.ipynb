```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ete3 import Tree, ProfileFace, TreeStyle, BarChartFace

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
```

```python
emapper = pd.read_csv("../Data/input/MM_jbrjl6fg.emapper.annotations.tsv", sep ="\t", comment = "#", header = None)
emapper.columns = ["query","seed_ortholog","evalue","score","eggNOG_OGs","max_annot_lvl","COG_category","Description","Preferred_name","GOs","EC","KEGG_ko","KEGG_Pathway","KEGG_Module","KEGG_Reaction","KEGG_rclass","BRITE","KEGG_TC","CAZy","BiGG_Reaction","PFAMs"]

OFOGs = pd.read_csv("../Data/input/Proteomes/OrthoFinder/Results_Dec03/Orthogroups/Orthogroups.GeneCount.tsv", sep ="\t", comment = "#", index_col = 0, header = 0)

OFOG_Annot = pd.DataFrame()
OFOG_Annot["OFOG"] = emapper["query"].str.split(".",expand=True)[0]
OFOG_Annot["eOG"] = emapper["eggNOG_OGs"].str.split("@",expand=True)[0]
```

```python
WLP_genes = pd.read_csv("../Data/input/MethanoAcetogenesisGenes.tsv", sep ="\t", comment = "#", header = None)
WLP_genes.columns = ["KOG","name","COG"]
WLP_genes = WLP_genes.explode('COG')
WLP_genes = WLP_genes.groupby(['COG']).agg({'name': lambda x: x.tolist()},axis=1)

OrderedCOGs = list(pd.read_csv("../Data/input/MethanoAcetogenesisGenes.tsv", sep ="\t", comment = "#", header = None).loc[:,2].unique())

WLP_PresGenes = WLP_genes.merge(OFOG_Annot.loc[:,["OFOG","eOG"]].set_index("eOG"), how="left", left_index=True, right_on="eOG")#, right_index=False)#.dropna()
COGs_KEGG = WLP_PresGenes.merge(OFOGs.reset_index(), how= "left", left_on="OFOG", right_on="Orthogroup").set_index("eOG").drop(["OFOG","name","Orthogroup","Total"], axis=1).fillna(0)
COGs_KEGG = COGs_KEGG.groupby(['eOG']).sum().loc[OrderedCOGs]
COGs_KEGG
```

```python
Root_OGMembers = pd.read_csv("../Data/input/EggNOG/1_members.tsv", sep ="\t", comment = "#", header = None, index_col = 1)
Root_OGMembers = Root_OGMembers[Root_OGMembers.index.isin(WLP_genes.index)]
Root_OGMembers["list"] = Root_OGMembers[4].str.split(",", expand = False)

WLP_PresGenes_EggNOG = WLP_genes.merge(Root_OGMembers["list"], how="left", left_on="COG", right_index=True)#, right_index)#, right_on=4)#.dropna()
WLP_PresGenes_EggNOG.name.values#.head()
```

```python
ExpandedOGs = WLP_PresGenes_EggNOG["list"].explode().str.split(".", expand = True, n=1).reset_index()

NCBIRefIDs = pd.read_csv("../Data/input/EggNOG/NCBIRefIDs.tsv", sep ="\t", comment = "#", header = None)
NCBIRefIDs.columns = ["TaxID", "TaxName"]
NCBIRefIDs = NCBIRefIDs.astype(str).set_index("TaxID")

for NCBIRefID in NCBIRefIDs.index:
    COGs_KEGG[NCBIRefID] = "NaN"
    for COG in COGs_KEGG.index:
        try: COGs_KEGG.loc[COG,NCBIRefID] = ExpandedOGs[ExpandedOGs.COG == COG][0].value_counts()[NCBIRefID]
        except: COGs_KEGG.loc[COG,NCBIRefID] = 0
            
COGs_KEGG
```

# Principle component analysis of the chosen species over their gene abundance in the aceto/methanogenic gene families

```python
COGs_boolean = COGs_KEGG[COGs_KEGG == 0.0].fillna(1)
COGs_boolean
```

```python
pipeline = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=2))])

pca_out = pipeline.fit_transform(COGs_KEGG)

plt.scatter(pca_out[:,0],pca_out[:,1])

#plt.legend(["Asgard","Acetogens","Methanogens","Outgroups"])
plt.xlabel('PCA1')# {round(pca.explained_variance_ratio_[0]*100.1)}%'))
plt.ylabel('PCA2')# {round(pca.explained_variance_ratio_[1]*100.1)}%'))


for i, txt in enumerate(COGs_KEGG.index):
    if pca_out[i,0] > 5:
        plt.annotate(txt, (pca_out[i,0],pca_out[i,1]))

plt.show()
```

```python
pipeline = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=2))])

pca_out = pipeline.fit_transform(COGs_boolean.transpose())

plt.scatter(pca_out[0:68,0],pca_out[0:68,1])#, color="r")
plt.scatter(pca_out[68:77,0],pca_out[68:77,1])#, color="r")
plt.scatter(pca_out[77:86,0],pca_out[77:86,1])#, color="r")
plt.scatter(pca_out[86:89,0],pca_out[86:89,1])#, color="r")
#plt.scatter(pca_out.components_[0][68:77], pca.components_[1][68:77])#, color="b")
#plt.scatter(pca_out.components_[0][77:86], pca.components_[1][77:86])#, color="g")
#plt.scatter(pca_out.components_[0][86:89], pca.components_[1][86:89])#, color="y")

plt.legend(["Asgard","Acetogens","Methanogens","Outgroups"])
plt.xlabel('PCA1')# {round(pca.explained_variance_ratio_[0]*100.1)}%'))
plt.ylabel('PCA2')
#plt.xlabel(str(f'PCA1 {round(pca.explained_variance_ratio_[0]*100.1)}%'))
#plt.ylabel(str(f'PCA2 {round(pca.explained_variance_ratio_[1]*100.1)}%'))
plt.show()
```

```python
SpeciesIndexDF_grouped = pd.read_csv("../Data/input/SpeciesIndexDF.tsv", sep ="\t", comment = "#", header = 0, index_col = "Proteome Id")["Organism"]
SpeciesIndexDF_grouped = SpeciesIndexDF_grouped.str.replace("Ca. ","").str.split(" ", n=1, expand = True)[0]
print(SpeciesIndexDF_grouped)

#manual correction
SpeciesIndexDF_grouped["UP000321408"] = "Lokiarchaeota"
SpeciesIndexDF_grouped["UP000625964"] = "Lokiarchaeota"
SpeciesIndexDF_grouped["UP000721749"] = "Heimdallarchaeota"
SpeciesIndexDF_grouped["UP000625964"] = "Lokiarchaeota"
SpeciesIndexDF_grouped["UP000721749"] = "Heimdallarchaeota"
SpeciesIndexDF_grouped["UP000775429"] = "Heimdallarchaeota"

GenusIndex = SpeciesIndexDF_grouped.append(NCBIRefIDs.TaxName.str.split(" ", n=1, expand = True)[0])

COGs_ByGenus = COGs_boolean.transpose().merge(GenusIndex, how = "left", left_index=True, right_index=True).groupby([0]).sum()#, left_on = index)#group
COGs_ByGenus_Boolean = COGs_ByGenus[COGs_ByGenus == 0.0].fillna(1).transpose()
COGs_ByGenus_Boolean
```

```python
pipeline = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=2))])

pca_out = pipeline.fit_transform(COGs_ByGenus_Boolean)

plt.scatter(pca_out[:,0],pca_out[:,1])

#plt.legend(["Asgard","Acetogens","Methanogens","Outgroups"])
plt.xlabel('PCA1')# {round(pca.explained_variance_ratio_[0]*100.1)}%'))
plt.ylabel('PCA2')# {round(pca.explained_variance_ratio_[1]*100.1)}%'))


for i, txt in enumerate(COGs_ByGenus_Boolean.index):
    if pca_out[i,0]:
        plt.annotate(txt, (pca_out[i,0],pca_out[i,1]), size = 5)

plt.show()
```

```python
pipeline = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=2))])

pca_out = pipeline.fit_transform(COGs_ByGenus_Boolean.transpose())

plt.scatter(pca_out[0:68,0],pca_out[0:68,1])#, color="r")
#plt.scatter(pca_out[68:77,0],pca_out[68:77,1])#, color="r")
#plt.scatter(pca_out[77:86,0],pca_out[77:86,1])#, color="r")
#plt.scatter(pca_out[86:89,0],pca_out[86:89,1])#, color="r")
#plt.scatter(pca_out.components_[0][68:77], pca.components_[1][68:77])#, color="b")
#plt.scatter(pca_out.components_[0][77:86], pca.components_[1][77:86])#, color="g")
#plt.scatter(pca_out.components_[0][86:89], pca.components_[1][86:89])#, color="y")

#plt.legend(["Asgard","Acetogens","Methanogens","Outgroups"])
plt.xlabel('PCA1')# {round(pca.explained_variance_ratio_[0]*100.1)}%'))
plt.ylabel('PCA2')
#plt.xlabel(str(f'PCA1 {round(pca.explained_variance_ratio_[0]*100.1)}%'))
#plt.ylabel(str(f'PCA2 {round(pca.explained_variance_ratio_[1]*100.1)}%'))


for i, txt in enumerate(COGs_ByGenus_Boolean.transpose().index):
    plt.annotate(txt, (pca_out[i,0],pca_out[i,1]), size = 6)
    
plt.show()
```

# Visualize it as a tree with ETE3

```python
SpeciesIndexDF = pd.read_csv("../Data/input/SpeciesIndexDF.tsv", sep ="\t", comment = "#", header = 0, index_col = "Proteome Id")["Organism"]
SpNamesID = pd.concat([SpeciesIndexDF,NCBIRefIDs["TaxName"]])
SpNamesID
```

```python
Combined = Tree("../Data/input/CombninedSpeciesTreeETE.nw", format=1)# made using orthofinder, NCBI ID's are slightly different to the ones from the strains in EggNOG so that's corrected in the .nw file.
Combined.ladderize()

for lf in Combined.iter_leaves():
    lf.add_features(profile = COGs_KEGG[lf.name].values)
    lf.add_features(deviation = [0 for x in range(len(COGs_KEGG))])
    lf.add_face(ProfileFace(max_v=36.0, min_v=0.0, center_v=10.0, style='heatmap', width=len(COGs_KEGG)*10, height=10), column=0, position="aligned")
    lf.name = SpNamesID[lf.name]
    
description = list(COGs_KEGG.index)
axisface = BarChartFace([0]*len(list(COGs_KEGG.index)), width=len(COGs_KEGG)*10, height=0, labels=list(COGs_KEGG.index), max_value=1, scale_fsize=1)

ts = TreeStyle()
ts.draw_guiding_lines = True
ts.show_leaf_name = True
ts.aligned_foot.add_face(axisface, 0)

#Combined.render('../Data/output/heatmap.png', tree_style=ts)
Combined.show(tree_style=ts)
```

# Hierarchical clustering

```python
from sklearn.metrics.pairwise import pairwise_distances
import seaborn as sns
import scipy.spatial as sp
import scipy.cluster.hierarchy as hc
```

```python
RootOGs = pd.read_pickle("../AllOFOG&eOGCounts.pkl")
RootOGs = RootOGs.loc[RootOGs.index.isin(WLP_genes.index)]
RootOGs
```

```python
RootOGs2 = RootOGs
WLP_genes.name = WLP_genes.name.apply(lambda x: '/'.join([str(i) for i in x])).str.replace(" ","")
WLP_genes.name["COG1150"] = "hdrC1/C2/D/E"
WLP_genes.name["COG1148"] = "hdrA1/A2"
WLP_genes.name["COG2048"] = "hdrB1/B2"
WLP_genes.name["COG4656"] = "metV/rnfC/C2"
WLP_genes.name["COG1145"] = "hydA2/fwdF/G"
WLP_genes.name["COG3383"] = "fdhA/B/F"
RootOGs2.index = WLP_genes.loc[RootOGs2.index].index + ': ' + WLP_genes.loc[RootOGs2.index].name
RootOGs2
```

```python
%%time
Jaccard_sim1 = pd.DataFrame(pairwise_distances(np.array(RootOGs), metric = "jaccard"), index=RootOGs.index, columns=RootOGs.index)
Jaccard_sim1
```
