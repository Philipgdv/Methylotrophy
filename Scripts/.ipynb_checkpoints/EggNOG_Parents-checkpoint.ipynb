```python
Bacteria = pd.read_csv("../Data/input/EggNOG/2.parents.tsv", sep ="\t", header = None)
Bacteria.columns = ["Bact","Root"]
Bacteria.Bact = Bacteria.Bact.str.replace("2:ENOG50","")
Bacteria.Root = Bacteria.Root.str.replace("1:ENOG50","")
Bacteria.Root = Bacteria.Root.str.split("|", expand = False)
Bacteria = Bacteria.explode('Root')
Bacteria = Bacteria.groupby(['Root']).agg({'Bact': lambda x: x.tolist()},axis=1)
Bacteria
```

Eukarya = pd.read_csv("../Data/input/EggNOG/2759.parents.tsv", sep ="\t", header = None)
Eukarya.columns = ["Euka","Root"]
Eukarya.Euka = Eukarya.Euka.str.replace("2759:ENOG50","")
Eukarya.Root = Eukarya.Root.str.replace("1:ENOG50","")
Eukarya.Root = Eukarya.Root.str.split("|", expand = False)
Eukarya = Eukarya.explode('Root')
Eukarya = Eukarya.groupby(['Root']).agg({'Euka': lambda x: x.tolist()},axis=1)
Eukarya

```python
Archaea = pd.read_csv("../Data/input/EggNOG/2157.parents.tsv", sep ="\t", header = None)
Archaea.columns = ["Arch","Root"]
Archaea.Arch = Archaea.Arch.str.replace("2157:ENOG50","")
Archaea.Root = Archaea.Root.str.replace("1:ENOG50","")
Archaea.Root = Archaea.Root.str.split("|", expand = False)
Archaea = Archaea.explode('Root')
Archaea = Archaea.groupby(['Root']).agg({'Arch': lambda x: x.tolist()},axis=1)
Archaea
```

```python
Root = Bacteria.merge(Archaea, how='inner', on='Root')#.merge(Eukarya, how='outer', on='Root')
Root = Root.loc[~Root.index.duplicated(keep='first'),:]
Root
```

```python
emapper = pd.read_csv("../Data/input/MM_u6x7q5tb.emapper.annotations.tsv", sep ="\t", comment = "#", header = None)
emapper = emapper.loc[:,[0,4]]
emapper.columns = ["OFOG","eOG"]
emapper.OFOG = emapper.OFOG.str.split(".",expand=True)[0]
emapper.eOG = emapper.eOG.str.split("@",expand=True)[0]
NewOGs = pd.read_csv("../Data/OFResults/Orthogroups.tsv", sep ="\t", header = 0)
emapper = NewOGs.merge(emapper, how='left', left_on="Orthogroup", right_on="OFOG").loc[:,["Orthogroup","eOG"]]
emapper
```

```python
i = 0
for OG in emapper[emapper.eOG.isna()].index:
    i += 1
    emapper.eOG.iloc[OG] = f"OFOG{i}"

emapper
```

```python
emapper = emapper.groupby(['eOG']).agg({'Orthogroup': lambda x: x.tolist()},axis=1)
emapper
```

```python
CrossMap = emapper.join(Root)
CrossMap = CrossMap.apply(lambda s: s.fillna({i: [] for i in CrossMap.index}))#.fillna({i: [] for i in CrossMap.index})#CrossMap.fillna("").apply(list)
CrossMap
```

```python
CrossMap["Gathered"] = "NaN"

for eOG in CrossMap.index:
    if eOG.startswith('arCOG'):
        CrossMap.Gathered[eOG] = [eOG] + CrossMap.Bact[eOG]# + CrossMap.Euka[eOG]
    elif eOG.startswith('COG'):
        CrossMap.Gathered[eOG] = [eOG] + CrossMap.Arch[eOG]# + CrossMap.Euka[eOG]
    elif eOG.startswith('KOG'):
        CrossMap.Gathered[eOG] = [eOG] + CrossMap.Arch[eOG] + CrossMap.Bact[eOG]
    else: 
        CrossMap.Gathered[eOG] = []
        
CrossMap
```

```python
import glob
http = urllib3.PoolManager()

try:
    os.mkdir(f"../Data/temp/")
except:
    print("Folder already exists") 

#http = urllib3.PoolManager()

for eOG in CrossMap.index:
    try: 
        os.mkdir(f"../Data/temp/{eOG}")
        print(f"Folder {eOG} is created")
    except:
        print(f"Folder {eOG} already exists")

    for OFOG in CrossMap.Orthogroup[eOG]:
        file = f'../Data/OFResults/MultipleSequenceAlignments/{OFOG}/{OFOG}.fa'
        shutil.copy(file, f'../Data/temp/{eOG}')
        print("Copied..." + OFOG)
    for EggNOG in CrossMap.Gathered[eOG]:
        try: 
            file = glob.glob(f"../Data/input/EggNOG/*/{EggNOG}.trimmed_alg.faa.gz")[0]
            shutil.copy(file, f'../Data/temp/{eOG}')
            print("Copied..." + EggNOG)
        except: 
            print(f"could not find {EggNOG}, download unaligned sequences:")
            file = f'http://eggnogapi5.embl.de/nog_data/text/fasta/{EggNOG}'
            r = http.request('GET',file)
            open(f'../Data/temp/{eOG}/{EggNOG}unalg.faa', "wb").write(r.data)
            print("Downloading..." + EggNOG)

    #with open(f"../Data/temp/{eOG}/targetOFOGs", 'w') as f:
    #    for line in CrossMap.Orthogroup[eOG]:
    #        f.write(f"{line}\n")
    #with open(f"../Data/temp/{eOG}/targetEggNOGs", 'w') as f:
    #    for line in CrossMap.Gathered[eOG]:
    #        if line != 'NaN':
    #            f.write(f"{line}\n")
```
