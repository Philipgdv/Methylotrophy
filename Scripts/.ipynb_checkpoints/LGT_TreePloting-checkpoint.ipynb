```python
from ete3 import PhyloTree, Tree, NCBITaxa, TreeStyle, AttrFace, NodeStyle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from os.path import join
import urllib3
import shutil
from Bio import SeqIO
import subprocess
import time

ncbi = NCBITaxa()
```

```python
COGs = pd.DataFrame(["COG0282","COG0190","COG4656","COG1410","COG1962","COG0407","COG5012"],
                    index = ["AckA","FolD","MetV","AcsE","MtrH","MtaA","MtaC"])
COGs
```

## Find all clades present in the selected trees

```python
ncbi.translate_to_names([49928])
```

```python
Attr_Clades_all = pd.Series()
for COG_name in COGs.index:
    COG = PhyloTree(f"../Data/LGTAnalysis/{COG_name}/{COG_name}.nwk", format=1, sp_naming_function=lambda name: ncbi.get_lineage(name.split('.')[0])[3])
    
    for lf in COG.iter_leaves():
        lf.add_feature("clade", ncbi.get_lineage(lf.name.split('.')[0])[3])

    clades = []
    for lf in COG.iter_leaves():
        clades.append(lf.clade)

    Attr_Clades_all = Attr_Clades_all.append(pd.Series(index=list(set(clades))))
    
Attr_Clades = pd.Series(index = Attr_Clades_all.index.unique())

for TaxID in Attr_Clades.index:
    Attr_Clades.loc[TaxID] = ncbi.translate_to_names([TaxID])[0]
```

## Treestyle Definition

```python
ts = TreeStyle()
ts.mode = "c"
ts.draw_guiding_lines = False
ts.show_leaf_name = False

colors = {1:"Black",
          49928:"RoyalBlue",
          2323:"RoyalBlue",
          1783275:"Coral",
          1935183:"OrangeRed",
          28890:"DarkOrange",
          2138240:"Aqua",
          2498710:"PaleTurquoise",
          508458:"Aquamarine",
          203691:"CadetBlue",
          3018035:"SteelBlue",
          40117:"MediumBlue",
          32066:"Navy",
          1224:"Indigo",
          2818505:"Lime",
          68297:"LimeGreen",
          200783:"ForestGreen",
          200918:"DarkOliveGreen",
          1783257:"Olive",
          200930:"Purple",
          1802340:"SlateBlue",
          1783270:"DarkViolet",
          1783272:"Plum",
          200938:"Fuchsia",
          200940:"DarkMagenta",
          29547:"RoyalBlue",
          1930617:"DarkKhaki",
          57723:"BlueViolet",
          2283796:"DarkSlateGrey"}

Colors_Clades = pd.DataFrame(colors.values(), index = colors.keys(), columns = ["colors"])
Colors_Clades["Names"] = Attr_Clades
```

# Reload the trees and color annotate them by clade, save to svg

```python
Attr_Clades_all = pd.Series()
for COG_name in COGs.index:
    COG = PhyloTree(f"../Data/LGTAnalysis/{COG_name}/{COG_name}.nwk", format=1, sp_naming_function=lambda name: ncbi.get_lineage(name.split('.')[0])[3])
    R = COG.get_midpoint_outgroup()
    COG.set_outgroup(R)
    COG.ladderize()
    
    for lf in COG.iter_leaves():
        lf.add_feature("clade", ncbi.get_lineage(lf.name.split('.')[0])[3])

    clades = []
    for lf in COG.iter_leaves():
        clades.append(lf.clade)
    
    for node in COG.traverse():
        if node.is_leaf()==False:
            leaf_clades=[leaf.clade for leaf in node.iter_leaves()]
            unique_clades = list(set(leaf_clades))
            node.add_feature("clade", 1)
            if (len(unique_clades)==1):
                node.clade = unique_clades[0]

    for node in COG.traverse():
        node.img_style['hz_line_color'] = colors[node.clade]
        node.img_style['vt_line_color'] = colors[node.clade]

    COG.render(f"../Data/LGTAnalysis/{COG_name}/{COG_name}.svg",tree_style=ts)
    #COG.show(tree_style=ts)
```
